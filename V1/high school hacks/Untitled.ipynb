```python
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
symbol_groups = list(chunks(stocks['Ticker'], 100))
symbol_strings = []
for i in range(0, len(symbol_groups)):
    symbol_strings.append(','.join(symbol_groups[i]))

final_dataframe = pd.DataFrame(columns=my_columns)
for symbol_string in symbol_strings:
    batch_api_call_url = f'https://www.alphavantage.co/batch?function=OVERVIEW&symbols={symbol_string}&apikey={api_key}'
    data_generic = requests.get(batch_api_call_url).json()
    for symbol in symbol_string.split(','):
        final_dataframe = final_dataframe.append(
            pd.Series(
                [
                    symbol,
                    data_generic[symbol]['Global Quote']['05. price'],
                    data_overview[symbol]['MarketCapitalization'],
                    'N/A',
                ],
                index=my_columns
            ),
            ignore_index=True
        )

final_dataframe
```
